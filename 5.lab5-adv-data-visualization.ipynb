### A seaborn jointplot - bivariate scatterplots and univariate histograms 

```python
# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, height=5)
```

```python
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris, height=8)

# histograms are a part of non-parametric data reduction methods
# two visualizations are not possible in matplotlib, but is present in seaborn.
```

### Seaborn's FacetGrid - to color the scatterplot by species

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species or by categories (i.e species here)

# functionalities like FacetGrid are not present in matplotlib
sns.FacetGrid(iris, hue="Species", height=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \ 
   .add_legend()

# in ....... \ map(plt.scatter, "attribute_as_x_axis", "attribute_as_x_axis") \ ......

# always infer from your plots, your projects depend on that

# we infer from the sepal length and width differences between the species
# that we can easily distinguish iris setosa from the other two categories with our naked eye by seeing the sepals

# certain points in the scatter plot are far from the rest
# are they outliers or does it only seem that way in the scatter plot?
# that can be checked by using box plots
```

```python
sns.FacetGrid(iris, hue="Species", height=5) \
   .map(plt.scatter, "PetalLengthCm", "PetalWidthCm") \
   .add_legend()

# remove extra white spaces, they might cause syntax errors due to use of indentation in python
```

- Boxplots can be used to:

### Seaborn Boxplot

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
```

```python
sns.boxplot(x="Species", y="PetalWidthCm", data=iris)

# we can also plot all four species in the same plot using seaborn
```

```python
# One way we can extend this plot is adding a layer of individual points on top of
# it through Seaborn's striplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines
# above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, linewidth=1)

# jitter is for data points to be put on the data
# here along with the box plot we can see the spread of the data, or what the data "looks like"
```

### Seaborn Violin plot

 - A violin plot combines the benefits of the previous two plots and simplifies them
 - Denser regions of the data are fatter, and sparser thiner in a violin plot

```python
ax = sns.violinplot(x="PetalLengthCm", data=iris)

# violin plot comes under which category of plots?
# it comes under univariate or bivariate plots?

# violin plots allow us to see where the data is denser or sparser, so help to visualiaze further
# box plots are just a box.
```

```python
ax = sns.violinplot(x="SepalWidthCm", data=iris)
```

```python
### Tune bandwidth
ax = sns.violinplot(x="SepalWidthCm", data=iris, bw=.15)
```

```python
ax = sns.violinplot(x="SepalWidthCm", data=iris, bw=.10)
```

```python
ax = sns.violinplot(x="SepalWidthCm", data=iris, bw=.04)

# bandwidth is specified to see how the data is actually set up, without smoothening of the data
```

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)

# see how one can observe the differences between iris setosa and the other two categories
# also see the similarities between iris virginica and versicolor
```

### kdeplot

```python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature

# kde full form: Kernel Density Estimate

sns.FacetGrid(iris, hue="Species", height=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()

# here the y axis has the density
```

### Pairplot

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
# 
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", height=3)

# in the plots shown below we can see that or 
# we can infer that the iris setosa can be distinguished from the other two categories

# pair plots cannot be used for visualization when we have many categories with lakhs of values
# heatmaps can used in those cases
```

```python
# The diagonal elements in a pairplot show the kde by default
# We can update these elements to show other things, such as a histogram
sns.pairplot(iris.drop("Id", axis=1), hue="Species", height=3, diag_kind="hist")
```

```python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species

# before using this, be clear with the understanding and representating the data

iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```
